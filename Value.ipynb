```python


def value_iteration(n, m, valid_actions, terminal, reward, gamma=0.95, delta=1e-6, max_iter=100000):
    S = n * m
    V = np.zeros(S, dtype=float)

    for it in range(max_iter):
        diff = 0.0

        for s in range(S):
            if terminal[s]:
                continue

            best = -1e18
            for a in valid_actions[s]:
                s2 = step_det(n, m, s, a)
                val = reward[s2] + (0.0 if terminal[s2] else gamma * V[s2])
                if val > best:
                    best = val

            diff = max(diff, abs(best - V[s]))
            V[s] = best

        if diff < delta:
            return V, it + 1, diff

    return V, max_iter, diff
```

```python
gamma = 0.2
delta = 1e-6   # seuil d'arrêt

V_vi, iters_vi, last_diff = value_iteration(
    n=7, m=7,
    valid_actions=valid_actions,
    terminal=terminal,
    reward=reward,
    gamma=gamma,
    delta=delta
)

print("✅ Value Iteration terminé")
print("Itérations:", iters_vi)
print("Diff finale:", last_diff)

print("\n--- Valeurs d'état V(s) (grille 7x7) ---")
print(np.round(V_vi.reshape(7,7), 3))
```
