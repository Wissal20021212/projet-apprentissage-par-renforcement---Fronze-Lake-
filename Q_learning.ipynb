```python
def epsilon_greedy(Q, s, valid_actions, epsilon):
    if random.random() < epsilon:
        return random.choice(valid_actions[s])
    else:
        # action qui maximise Q(s,a)
        return max(valid_actions[s], key=lambda a: Q[s,a])
```

```python
def q_learning(n, m, valid_actions, terminal, reward,
               alpha=0.1, gamma=0.95,
               epsilon=0.1,
               episodes=5000,
               max_steps=200):

    S = n*m
    Q = np.zeros((S, 4))  # 4 actions

    for ep in range(episodes):

        s = start_state  # on commence toujours au start

        for step in range(max_steps):

            # choisir action
            a = epsilon_greedy(Q, s, valid_actions, epsilon)

            # transition
            s2 = step_det(n, m, s, a)

            # récompense
            r = reward[s2]

            # mise à jour Q-learning
            best_next = max([Q[s2,a2] for a2 in valid_actions[s2]]) if not terminal[s2] else 0

            Q[s,a] = Q[s,a] + alpha * (r + gamma * best_next - Q[s,a])

            s = s2

            if terminal[s]:
                break

    return Q
```

```python
alpha = 0.1
gamma = 0.95
epsilon = 0.5

Q_ql = q_learning(
    n=7, m=7,
    valid_actions=valid_actions,
    terminal=terminal,
    reward=reward,
    alpha=alpha,
    gamma=gamma,
    epsilon=epsilon,
    episodes=5000
)

print("Q-table obtenue :")
print(np.round(Q_ql, 3))
policy_ql = np.zeros(49, dtype=int)

for s in range(49):
    if not terminal[s]:
        policy_ql[s] = max(valid_actions[s], key=lambda a: Q_ql[s,a])
plot_policy_grid_clean(
    n=7,
    m=7,
    policy=policy_ql,
    holes=holes,
    goal_state=goal_state,
    start_state=start_state,
    title="Q-learning – Politique finale"
) 
```
